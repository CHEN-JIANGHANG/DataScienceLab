```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
plt.rc("figure", figsize=(10, 6))
np.set_printoptions(precision=4)

colors = ["#348ABD", "#A60628", "#7A68A6", "#467821", "#D55E00",
          "#CC79A7", "#56B4E9", "#009E73", "#F0E442", "#0072B2"]

import warnings
warnings.simplefilter('ignore', DeprecationWarning)
```

###Load Data

```python
df = pd.read_csv('../data/critics.csv')
# df.head(2)
df.count()
```

###Train-Test-Split
[back to top](#contents)

```python
from sklearn.cross_validation import train_test_split

X = df.quote
y = (df.fresh == 'fresh').values.astype(np.int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("Train data shape: %r, Train target shape: %r" % (X_train.shape, y_train.shape))
print("Test data shape: %r, Test target shape: %r" % (X_test.shape, y_test.shape))
```

###Vectorizer
[back to top](#contents)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=2)

X_train_vec = vectorizer.fit_transform(X_train)
y_train = y_train

n_samples, n_features = X_train_vec.shape
print("No. of samples = %d" % n_samples)
print("No. of features = %d" % n_features)
```

###PCA evaluation

###Choose Estimator
[back to top](#contents)

```python
%%time

from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV

from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()

pipeline = Pipeline([
    ('vec', TfidfVectorizer(min_df=1, max_df=0.8, use_idf=True)),
    ('clf', clf)
])

clf = pipeline.fit(X_train, y_train)

train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)

print "Train Score: %0.2f%%" % (100 * (train_score))
print "Test Score: %0.2f%%" % (100 * (test_score))
```

###Cross-validation Iterator
[back to top](#contents)

```python
%%time

from sklearn.cross_validation import ShuffleSplit
from sklearn.cross_validation import cross_val_score

n_subsamples = 1000
X_small_train, y_small_train = X_train[:n_subsamples], y_train[:n_subsamples]

cv = ShuffleSplit(n_samples, 
                  n_iter=50,
                  test_size=0.2,
                  random_state=0)

scores = cross_val_score(pipeline, X, y, 
                         cv=cv, 
                         n_jobs=2,
                         scoring='accuracy')

print "CV Score (min): %0.2f%%" % (100 * scores.min())
print "CV Score (mean): %0.2f%%" % (100 * scores.mean())
print "CV Score (max): %0.2f%%" % (100 * scores.max())
```

```python
from scipy.stats.kde import gaussian_kde
_ = plt.hist(scores, range=(0, 1), bins=30, alpha=0.2)
x = np.linspace(0, 1, 1000)
smoothed = gaussian_kde(scores).evaluate(x)
plt.plot(x, smoothed, label="Smoothed distribution")
top = np.max(smoothed)
plt.vlines([np.mean(scores)], 0, top, color='r', label="Mean test score")
plt.vlines([np.median(scores)], 0, top, color='b', linestyles='dashed',
           label="Median test score")
plt.legend(loc='best')
_ = plt.title("Cross Validated Test scores distribution")
```

####Learning Curve
[back to top](#contents)

```python
def LearningCurve(model, title, X, y):
    plt.figure()
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(model, X, y)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, 
                     train_scores_mean + train_scores_std, alpha=0.1,color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",label="Cross-validation score")
    plt.title(title)
    plt.legend(loc="best")
    return plt
```

```python
from sklearn.learning_curve import learning_curve

alpha = gs.best_params_['clf__alpha']
estimator = MultinomialNB(alpha=alpha)
title = "Learning Curves (Multinomial NB, alpha=%.4f)" % (alpha)
LearningCurve(estimator, title, X_train_vec, y_train);
```

####Validation Curve
[back to top](#contents)

```python
from sklearn.learning_curve import validation_curve

alpha_range = np.logspace(-6, 0, 7)

train_scores, validation_scores = validation_curve(MultinomialNB(),
                                                   X_train_vec,
                                                   y_train,
                                                   'alpha',
                                                   alpha_range,
                                                   cv=5,
                                                   n_jobs=-1
                                                   )

plt.semilogx(alpha_range, train_scores.mean(axis=1), 'o-', color="r", label='Training score')
plt.semilogx(alpha_range, validation_scores.mean(axis=1), 'o-', color="g", label='Cross-validation score')
plt.legend(loc='best')
plt.xlabel("Alpha Range")
_ = plt.title('Validation curves');
```

####Confusion Matrix
[back to top](#contents)

```python
from sklearn.metrics import confusion_matrix

def plot_confusion(cm):
    plt.imshow(cm, interpolation='nearest')
    plt.title('Confusion matrix')
    plt.set_cmap('Blues')
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

y_pred = gs.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm_normalized = cm.astype(np.float64) / cm.sum(axis=1)

plot_confusion(cm_normalized)
# print cm_normalized
```
