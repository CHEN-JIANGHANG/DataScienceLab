###Preliminaries

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
plt.rc("figure", figsize=(10, 6))
np.set_printoptions(precision=4)

colors = ["#348ABD", "#A60628", "#7A68A6", "#467821", "#D55E00",
          "#CC79A7", "#56B4E9", "#009E73", "#F0E442", "#0072B2"]

import warnings
warnings.simplefilter('ignore', DeprecationWarning)
```

###Load Data
[back to top](#contents)

```python
un = pd.read_csv('../data/un.csv')
un.head(2)
# un.count()
```

```python
df = un
features = ['lifeMale', 'lifeFemale', 'infantMortality', 'GDPperCapita']
target = 'region'

df_features = df[features]
df_features = df_features.fillna(-1) #to indicate -1 is missing value
# df_features = df_features.fillna(df.dropna().median())

# df_features.head()
df_features.count()
```

```python
X = df_features.values
y = df[target].values
n_samples, n_features = X.shape
print("No. of samples = %d" % n_samples)
print("No. of features = %d" % n_features)
```

```python
sns.pairplot(df_features, size=2.5);
```

###Train-Test-Split
[back to top](#contents)

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print "Train data shape: %r, Train target shape: %r" % (X_train.shape, y_train.shape)
print "Test data shape: %r, Test target shape: %r" % (X_test.shape, y_test.shape)
```

###K-Means

```python
cluster = 3 #estimate clusters
```

```python
from sklearn.cluster import KMeans

est = KMeans(cluster)
est.fit(X_train)

predicted = est.predict(X_train)

cmap = sns.diverging_palette(220, 10, as_cmap=True)
plt.scatter(X_train[:, 3], X_train[:, 2], c=predicted, s=50, 
            cmap=cmap)
plt.xlim(0)
plt.ylim(0)
plt.xlabel(df_features.columns[3])
plt.ylabel(df_features.columns[2]);
```

###Check with Elbow Method

```python
from scipy.cluster.vq import kmeans,vq
from scipy.spatial.distance import cdist

K = range(1,10)
KM = [kmeans(X_train,k) for k in K]
centroids = [cent for (cent,var) in KM]
D_k = [cdist(X_train, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]
avgWithinSS = [sum(d)/X_train.shape[0] for d in dist] 
```

```python
kIdx = cluster - 1
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(K, avgWithinSS, '*--')
ax.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
        markeredgewidth=2, markeredgecolor='r', 
        markerfacecolor='None')
plt.grid(True)
plt.xlabel('No. of clusters')
plt.ylabel('Average within-cluster sum of squares')
tt = plt.title('Elbow for K-Means clustering') 
```
