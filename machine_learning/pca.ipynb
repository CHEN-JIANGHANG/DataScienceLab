###Preliminaries
[back to top](#contents)

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
plt.rc("figure", figsize=(10, 6))
np.set_printoptions(precision=4)

colors = ["#348ABD", "#A60628", "#7A68A6", "#467821", "#D55E00",
          "#CC79A7", "#56B4E9", "#009E73", "#F0E442", "#0072B2"]
cmap = sns.diverging_palette(220, 10, as_cmap=True)

import warnings
warnings.simplefilter('ignore', DeprecationWarning)
```

###Load Data
[back to top](#contents)

```python
from sklearn.datasets import load_digits
digits = load_digits()

X = digits.data
y = digits.target
```

###Train-Test-Split
[back to top](#contents)

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print "Train data shape: %r, Train target shape: %r" % (X_train.shape, y_train.shape)
print "Test data shape: %r, Test target shape: %r" % (X_test.shape, y_test.shape)
```

###PCA
[back to top](#contents)

```python
# from sklearn.decomposition import TruncatedSVD
# pca = TruncatedSVD(n_components=2)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)

print X_train.shape
print X_pca.shape
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, edgecolor='none', alpha=0.5,
            cmap=cmap)
plt.colorbar();
```

###Choose Number of Components
[back to top](#contents)

```python
pca = PCA().fit(X_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No. of components')
plt.ylabel('Cumulative explained variance');
```
