```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")

from scipy import stats

# sns.set(rc={"figure.figsize": (6, 6)})
```

###Load Datasets
[back to top](#contents)

```python
tips = pd.read_csv('data/tips.csv')
iris = pd.read_csv('data/iris.csv')
titanic = pd.read_csv('data/titanic.csv')
```

```python
tips.head()
```

```python
iris.head()
```

```python
def woman_child_or_man(passenger):
    age, sex = passenger
    if age < 16:
        return "child"
    else:
        return dict(male="man", female="woman")[sex]

titanic["who"] = titanic[["age", "sex"]].apply(woman_child_or_man, axis=1)
titanic.head()
```

###Set Palettes

```python
pal_titanic = dict(man="#4682B4", woman="#CD5C5C", child="#2E8B57", male="#6495ED", female="#F08080")
pal_tips = dict(male="#6495ED", female="#F08080")
pal_iris = dict(setosa="#4682B4", versicolor="#CD5C5C", virginica="#2E8B57")
```

###Factorplot

```python
df = titanic
data1 = "class" #Categorical

sns.factorplot(data1, data=df, palette="PuBu_d");
```

```python
df = titanic
data1 = "survived" #Categorical
data2 = "sex" #Categorical - count

sns.factorplot(data1, data=df, hue=data2, palette="PuBu_d");
```

```python
df = titanic
data1 = "deck" #Categorical
data2 = "fare" #Numerical

sns.factorplot(data1, data2, data=df, palette="PuBu_d");
```

```python
df = titanic
data1 = "class" #Categorical
data2 = "survived" #Categorical -  counting probability

sns.factorplot(data1, data2, data=df).set(ylim=(0, 1));
```

```python
df = titanic
data1 = "class" #Categorical
data2 = "survived" #Categorical
data3 = "sex" #Categorical

sns.factorplot(data1, data2, data=df, hue=data3, palette=pal_titanic).set(ylim=(0, 1));
```

###Histograms

```python
sns.set_style("white")
plt.hist(tips.total_bill, color="#6495ED", alpha=.5);
```

```python
data1 = iris.sepal_length
data2 = iris.sepal_width

max_data = np.r_[data1, data2].max()
bins = np.linspace(0, max_data, max_data + 1)
plt.hist(data1, bins, normed=True, color="#6495ED", alpha=.5)
plt.hist(data2, bins, normed=True, color="#F08080", alpha=.5);
```

###Hexbin Plot

```python
x = tips.total_bill
y = tips.size
with sns.axes_style("white"):
    sns.jointplot(x, y, kind="hex");
```

###Distribution Plots

```python
fg = sns.FacetGrid(titanic, hue="sex", aspect=3, palette=pal)
fg.map(sns.kdeplot, "age", shade=True)
fg.set(xlim=(0, 80))
plt.legend(loc='best');
```

```python
fg = sns.FacetGrid(titanic, hue="who", aspect=3, palette=pal)
fg.map(sns.kdeplot, "age", shade=True)
fg.set(xlim=(0, 80))
plt.legend(loc='best');
```

```python
fg = sns.FacetGrid(titanic, hue="class", aspect=3, palette="BuPu_d")
fg.map(sns.kdeplot, "age", shade=True)
fg.set(xlim=(0, 80))
plt.legend(loc='best');
```

```python
fg = sns.FacetGrid(titanic, col="sex", row="class", hue="sex", size=2.5, aspect=2.5, palette=pal)
fg.map(sns.kdeplot, "age", shade=True)
fg.map(sns.rugplot, "age")
sns.despine(left=True)
fg.set(xlim=(0, 80));
```

```python
data = tips.tip
sns.set_palette("deep")
sns.set_style("white")
fig = plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
sns.kdeplot(data, shade=True)
plt.title('kde')

plt.subplot(2, 2, 2)
sns.distplot(data)
plt.title('displot')

plt.subplot(2, 2, 3)
sns.distplot(data, rug=True, hist=False)
plt.title('displot-rug')

plt.subplot(2, 2, 4)
sns.distplot(data, kde=False, fit=stats.norm)
plt.title('displot_normal')

plt.tight_layout();
```

```python
sns.distplot(data,
             kde_kws={"color": "seagreen", "lw": 3, "label": "KDE"},
             hist_kws={"histtype": "stepfilled", "color": "slategray"});
```

```python
plt.figure(figsize=(4, 7))
sns.distplot(data, color="dodgerblue", vertical=True);
```

###Factorplot

```python
fg = sns.factorplot("class", "survived", data=titanic, hue="who", col="who", palette=pal, aspect=.4)
fg.set(ylim=(0, 1))
fg.despine(left=True)
```

```python
fg = sns.factorplot("adult_male", "survived", data=titanic, col="class", hue="class",
                    aspect=.33, palette="BuPu_d")
fg.set(ylim=(0, 1))
fg.despine(left=True);
```

###Boxplot and Violinpolot
- Boxplot: compare 2 or more distributions

```python
data1 = tips.total_bill #distribution
data2 = tips.day #category

sns.boxplot(data1, data2);
```

```python
data1 = titanic.age #distribution
data2 = titanic.survived #category

sns.boxplot(data1, data2);
```

```python
data1 = titanic.fare #distribution
data2 = titanic.pclass #category

sns.violinplot(data1, data2, color="BuPu_d").set_ylim(0, 600)
sns.despine(left=True);
```

###Set Styles

```python
data = tips.tip
sns.set_palette("deep")

fig = plt.figure(figsize=(8, 6))
plt.subplot(3, 2, 1)
sns.set_style("darkgrid")
sns.kdeplot(data, shade=True)
plt.title('darkgrid - default')

plt.subplot(3, 2, 2)
sns.set_style("dark")
sns.kdeplot(data, shade=True)
plt.title('dark')

plt.subplot(3, 2, 3)
sns.set_style("white")
sns.kdeplot(data, shade=True)
plt.title('white')

plt.subplot(3, 2, 4)
sns.set_style("ticks")
sns.kdeplot(data, shade=True)
plt.title('ticks')

plt.subplot(3, 2, 5)
sns.set_style("whitegrid")
sns.kdeplot(data, shade=True)
plt.title('whitegrid')

fig.tight_layout();
```

```python
###Set Palettes
```

```python
#deep, muted, bright, pastel, dark, colorblind, hls, husl
data = tips.tip

fig = plt.figure(figsize=(8, 6))
sns.set_style("white")

plt.subplot(3, 2, 1)
sns.set_palette("deep")
sns.kdeplot(data, shade=True)
plt.title('deep')

plt.subplot(3, 2, 2)
sns.set_palette("muted")
sns.kdeplot(data, shade=True)
plt.title('muted')

plt.subplot(3, 2, 3)
sns.set_palette("bright")
sns.kdeplot(data, shade=True)
plt.title('bright')

plt.subplot(3, 2, 4)
sns.set_palette("dark")
sns.kdeplot(data, shade=True)
plt.title('dark')

plt.subplot(3, 2, 5)
sns.set_palette("cubehelix")
sns.kdeplot(data, shade=True)
plt.title('cubehelix')

plt.subplot(3, 2, 6)
sns.set_palette("husl")
sns.kdeplot(data, shade=True)
plt.title('husl')

fig.tight_layout();
```
