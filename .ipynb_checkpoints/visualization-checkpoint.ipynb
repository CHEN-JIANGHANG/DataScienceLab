#Scatter-Matrix

```python
s = df.iloc[:,0:3] #set number of columns

g = sns.PairGrid(s, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```
