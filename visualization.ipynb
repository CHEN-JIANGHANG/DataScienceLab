###Set Styles
[back to top](#contents)

```python
data = tips.tip
sns.set_palette("deep")

fig = plt.figure(figsize=(8, 6))
plt.subplot(3, 2, 1)
sns.set_style("darkgrid")
sns.kdeplot(data, shade=True)
plt.title('darkgrid - default')

plt.subplot(3, 2, 2)
sns.set_style("dark")
sns.kdeplot(data, shade=True)
plt.title('dark')

plt.subplot(3, 2, 3)
sns.set_style("white")
sns.kdeplot(data, shade=True)
plt.title('white')

plt.subplot(3, 2, 4)
sns.set_style("ticks")
sns.kdeplot(data, shade=True)
plt.title('ticks')

plt.subplot(3, 2, 5)
sns.set_style("whitegrid")
sns.kdeplot(data, shade=True)
plt.title('whitegrid')

fig.tight_layout();
```

###Histograms

```python
sns.set_style("darkgrid")
plt.hist(tips.total_bill);
```

```python
data1 = iris.sepal_length
data2 = iris.sepal_width

max_data = np.r_[data1, data2].max()
bins = np.linspace(0, max_data, max_data + 1)
plt.hist(data1, bins, normed=True, color="#6495ED", alpha=.5)
plt.hist(data2, bins, normed=True, color="#F08080", alpha=.5);
```

```python
plt.hist(tips.total_bill, 20, histtype="stepfilled", alpha=.7);
```

###Hexbin Plot

```python
x = tips.total_bill
y = tips.size
with sns.axes_style("white"):
    sns.jointplot(x, y, kind="hex");
```

###Distribution Plots

```python
data = tips.tip
sns.set_palette("deep")
sns.set_style("white")
fig = plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
sns.kdeplot(data, shade=True)
plt.title('kde')

plt.subplot(2, 2, 2)
sns.distplot(data)
plt.title('displot')

plt.subplot(2, 2, 3)
sns.distplot(data, rug=True, hist=False)
plt.title('displot-rug')

plt.subplot(2, 2, 4)
sns.distplot(data, kde=False, fit=stats.norm)
plt.title('displot_normal')

plt.tight_layout();
```

```python
sns.distplot(data,
             kde_kws={"color": "seagreen", "lw": 3, "label": "KDE"},
             hist_kws={"histtype": "stepfilled", "color": "slategray"});
```

```python
plt.figure(figsize=(4, 7))
sns.distplot(data, color="dodgerblue", vertical=True);
```
